## 득표수 구하기

> 주어진 리스트는 반장 선거 투표 결과이다. 이영희의 총 득표수를 출력하시오.

```
[출력 예시]
4
```

````python
students = ['이영희', '김철수', '이영희', '조민지', '김철수', '조민지', '이영희', '이영희']

# 아래에 코드를 작성하시오.
votes = 0
for name in students:
  if name == '이영희':
    votes += 1
print(votes)
````

## 최댓값 구하기

> 주어진 리스트의 요소 중에서 최댓값을 출력하시오.

---
```
[출력 예시]
22
```

````python
numbers = [7, 10, 22, 4, 3, 17]

# 아래에 코드를 작성하시오.0
print(max(numbers))
max_value = 0
for num in numbers:
  if num > max_value:
    max_value = num
print(max_value)
````

## 최댓값과 등장 횟수 구하기

```

```

````python
numbers = [7, 10, 22, 7, 22, 22]

# 아래에 코드를 작성하시오.
M = max(numbers)
total = 0
for a in numbers:
  if a == M:
    total += 1
print(M, total)

max_value = 0
for num in numbers:
  if num > max_value:
    max_value = num
    count = 1
  elif num == max_value:
    count +=1
print(max_value, count)
  
````

## 5의 개수 구하기

## 'a'가 싫어


> 입력으로 짧은 영단어 word가 주어질 때, 해당 단어에서 'a'를 모두 제거한 결과를 출력하시오.

---
```
[입력 예시]
apple

[출력 예시]
pple
```

````python
word = input()

# 아래에 코드를 작성하시오.
print(word.replace('a',''))

result = ''
for char in word:
  if char != 'a':
    result += char
print(result)
````

## 단어 뒤집기

> 입력으로 짧은 영어단어 word가 주어질 때, 해당 단어를 역순으로 뒤집은 결과를 출력하시오.

---
```
[입력 예시]
apple

[출력 예시]
elppa
```

````python
word = input()

# 아래에 코드를 작성하시오.
print(word[::-1])

re = ''
for char in word:
  re = char + re
print(re)
````
